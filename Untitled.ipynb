```python
def openlist(text):
    
    returnObj = []
    with open(text, "r") as f:
        for line in f:
            returnObj.append(float(line[:-1]))
    
    return returnObj
```

```python
cos = openlist("cos_clip.txt")
dot = openlist("dot_clip.txt")
eu = openlist("eu_clip.txt")
```

```python
import matplotlib.pyplot as plt
```
